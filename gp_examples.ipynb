```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import os

# 1. Generate data
np.random.seed(42)
n_points = 500
x = np.random.normal(0, 1, n_points)
y = np.random.normal(0, 1, n_points)

# 2. Calculate distance from mean (0, 0)
distances = np.sqrt(x**2 + y**2)

# 3. Create figure
fig, ax = plt.subplots(figsize=(6, 6))

# 4. Scatter plot with colors based on distance
# Use reversed colormap (dark purple near mean, light purple far away)
scatter = ax.scatter(x, y, c=distances, cmap='RdPu_r', alpha=0.7, s=50, edgecolors='none')

# 5. Marginal distributions
x_axis = np.linspace(-5, 5, 500)
y_axis = np.linspace(-5, 5, 500)

pdf_x = norm.pdf(x_axis, 0, 1) * 2.5 - 5
pdf_y = norm.pdf(y_axis, 0, 1) * 2.5 - 5

# Bottom curve (along X1)
ax.plot(x_axis, pdf_x, color='gray', linestyle='--')

# Left curve (along X2)
ax.plot(pdf_y, y_axis, color='gray', linestyle='--')

# 6. Styling
ax.set_xlabel(r'$f(\mathbf{a})$', fontsize=12)
ax.set_ylabel(r"$f(\mathbf{a}')$", fontsize=12)
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_xticks([-4, -2, 0, 2, 4])
ax.set_yticks([-4, -2, 0, 2, 4])

plt.tight_layout()

save_folder = "plots"
if not os.path.exists(save_folder):
    os.makedirs(save_folder)

filename = f"bivariate_normal_n{n_points}.png"
filepath = os.path.join(save_folder, filename)

plt.savefig(filepath, dpi=300, bbox_inches='tight')
print(f"Plot gespeichert unter: {filepath}")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# Set random seed for reproducibility
# np.random.seed(42)

# 1. Parameters for distribution
n_points = 200
mean = [0, 0]
# Change covariance here (0.8 creates strong positive correlation)
covariance = 0
cov_matrix = [[1, covariance], 
              [covariance, 1]]
# 2. Generate data
data = np.random.multivariate_normal(mean, cov_matrix, n_points)
x1, x2 = data[:, 0], data[:, 1]

# 3. Calculate individual distances for ALL points (for normalization)
all_distances = np.concatenate([np.abs(x1), np.abs(x2)])
dist_min = all_distances.min()
dist_max = all_distances.max()


# 3. Create plot
fig, ax = plt.subplots(figsize=(8, 8))

# Colors based on distance (reversed: dark near mean, light far away)

# 4. Plot points and connections
for i in range(n_points):
    # Line between x1 (position 0) and x2 (position 1)

    dist_x1 = np.abs(x1[i])
    dist_x2 = np.abs(x2[i])
    
    # Normalize each distance
    norm_dist_x1 = (dist_x1 - dist_min) / (dist_max - dist_min) if dist_max > dist_min else 0
    norm_dist_x2 = (dist_x2 - dist_min) / (dist_max - dist_min) if dist_max > dist_min else 0
    
    # Get individual colors
    color_x1 = cm.RdPu_r(norm_dist_x1)
    color_x2 = cm.RdPu_r(norm_dist_x2)
    
    ax.plot([0, 1], [x1[i], x2[i]], color="black", alpha=0.3, linewidth=1)
    
    # Points at the ends
    ax.scatter(0, x1[i], color=color_x1 , alpha=0.6, s=40)
    ax.scatter(1, x2[i], color=color_x2, alpha=0.6, s=40)

# 5. Styling
ax.set_xticks([0, 1])
ax.set_xticklabels(['$f(\mathbf{a}_1)$', '$f(\mathbf{a}_2)$'])
ax.set_ylabel('Observed value', fontsize=12)
ax.set_xlabel('Random variable', fontsize=12)
# ax.set_title(f'Covariance: {covariance}')
ax.set_xlim([-0.2, 1.2])
ax.set_ylim([-3, 3])

plt.tight_layout()
filename = f"connected_lines{n_points}.png"
filepath = os.path.join(save_folder, filename)

plt.savefig(filepath, dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import ipywidgets as widgets
from IPython.display import display

def plot_variables(n_variables=4, covariance=0.3, n_points=50):
    # 1. Mittelwert und Kovarianzmatrix
    mean = [0] * n_variables
    cov_matrix = np.eye(n_variables) * (1 - covariance) + covariance
    
    # 2. Daten generieren
    data = np.random.multivariate_normal(mean, cov_matrix, n_points)
    
    # 3. Normalisierung fr die Farben (basierend auf dem Betrag aller Werte)
    all_values_abs = np.abs(data)
    dist_min = all_values_abs.min()
    dist_max = all_values_abs.max()

    # 4. Plot erstellen
    fig, ax = plt.subplots(figsize=(12, 6))
    
    for i in range(min(100, n_points)):  # Begrenzung fr Performance
        y_values = data[i, :]
        x_positions = np.arange(n_variables)
        
        # Linien zwischen den Variablen (hier nutzen wir eine neutrale Farbe)
        ax.plot(x_positions, y_values, color="black", alpha=0.3, linewidth=0.8, zorder=1)
        
        # Punkte individuell einfrben
        for j in range(n_variables):
            val = y_values[j]
            dist = np.abs(val)
            
            # Normalisierung des Abstands fr die Farbwahl
            norm_dist = (dist - dist_min) / (dist_max - dist_min) if dist_max > dist_min else 0
            
            # Farbe aus RdBu whlen (oder RdBu_r fr umgekehrt)
            point_color = cm.RdPu_r(norm_dist)
            
            ax.scatter(x_positions[j], val, color=point_color, alpha=0.6, s=25, zorder=2)
    
    # 5. Styling
    ax.set_ylabel('Observed value', fontsize=12)
    ax.set_xlabel('Random Variable', fontsize=12)
    # ax.set_title(f'{n_variables} Variablen | Kovarianz: {covariance:.2f})')
    ax.set_xlim([-0.5, n_variables - 0.5])
    ax.set_ylim([-3.5, 3.5])
    
    # Dynamische Ticks
    if n_variables <= 10:
        tick_indices = list(range(n_variables))
    elif n_variables <= 50:
        tick_indices = list(range(0, n_variables, 10))
    else:
        tick_indices = list(range(0, n_variables, 20))
    
    ax.set_xticks(tick_indices)
    ax.set_xticklabels([f'$f(\mathbf{{a}}_{{{i+1}}})$' for i in tick_indices])

    plt.tight_layout()
    filename = f"connected_all_lines{n_points}.png"
    filepath = os.path.join(save_folder, filename)

    plt.savefig(filepath, dpi=300, bbox_inches='tight')
    plt.show()

# Interaktive Widgets
widgets.interact(plot_variables,
                 n_variables=widgets.IntSlider(min=2, max=100, step=1, value=50),
                 covariance=widgets.FloatSlider(min=0, max=0.99, step=0.05, value=0.3),
                 n_points=widgets.IntSlider(min=5, max=100, step=5, value=100))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import ipywidgets as widgets
from IPython.display import display

np.random.seed(42)  # Fr Reproduzierbarkeit

def plot_gp_samples_interactive(n_variables=100, length_scale=10.0, n_series=10):
    # X-axis (variable indices)
    x_ax = np.arange(n_variables).reshape(-1, 1)
    
    # Kernel function (Squared Exponential / RBF)
    sq_dist = np.sum(x_ax**2, 1).reshape(-1, 1) + np.sum(x_ax**2, 1) - 2 * np.dot(x_ax, x_ax.T)
    cov_matrix = np.exp(-0.5 * (1/length_scale**2) * sq_dist)
    
    # Add nugget for numerical stability
    cov_matrix += 1e-8 * np.eye(n_variables)
    
    # Generate data (mean = 0)
    mean = np.zeros(n_variables)
    data = np.random.multivariate_normal(mean, cov_matrix, n_series)
    
    # 3. Normalisierung fr die Farben (basierend auf dem Betrag aller Werte)
    all_values_abs = np.abs(data)
    dist_min = all_values_abs.min()
    dist_max = all_values_abs.max()
    # Plot
    fig, ax = plt.subplots(figsize=(12, 6))

    for i in range(min(100, n_series)):  # Begrenzung fr Performance
        y_values = data[i, :]
        x_positions = np.arange(n_variables)
        
        # Linien zwischen den Variablen (hier nutzen wir eine neutrale Farbe)
        ax.plot(x_positions, y_values, color="black", alpha=0.3, linewidth=0.8, zorder=1)
        for j in range(n_variables):
            val = y_values[j]
            dist = np.abs(val)
            
            # Normalisierung des Abstands fr die Farbwahl
            norm_dist = (dist - dist_min) / (dist_max - dist_min) if dist_max > dist_min else 0
            
            # Farbe aus RdBu whlen (oder RdBu_r fr umgekehrt)
            point_color = cm.RdPu_r(norm_dist)
            
            ax.scatter(x_positions[j], val, color=point_color, alpha=0.6, s=25, zorder=2)
    # Styling
    ax.set_xlabel('Random Variable', fontsize=12)
    ax.set_ylabel('Observed value', fontsize=12)
    
    # Adjust ticks based on n_variables
    if n_variables <= 10:
        tick_indices = list(range(n_variables))
    elif n_variables <= 50:
        tick_indices = list(range(0, n_variables, 10))
    else:
        tick_indices = list(range(0, n_variables, 20))
    
    ax.set_xticks(tick_indices)
    ax.set_xticklabels([f'$f(\mathbf{{a}}_{{{i+1}}})$' for i in tick_indices])
    ax.set_ylim([-3, 3])
    plt.tight_layout()
    filename = f"funktion_all_lines{n_variables}with_dots.png"
    # filename = f"funktion_all_lines{n_variables}.png"
    filepath = os.path.join(save_folder, filename)

    plt.savefig(filepath, dpi=300, bbox_inches='tight')
    plt.show()

# Interactive widgets
widgets.interact(plot_gp_samples_interactive,
                 n_variables=widgets.IntSlider(min=2, max=200, step=1, value=100, 
                                               description='Variables'),
                 length_scale=widgets.FloatSlider(min=0.5, max=50.0, step=0.5, value=10.0,
                                                  description='Length Scale'),
                 n_series=widgets.IntSlider(min=1, max=1000, step=1, value=10,
                                           description='Series'))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import ipywidgets as widgets
from scipy.stats import norm

def plot_gp_constrained(n_variables=100, length_scale=10.0, n_series=10, const_idx=10, const_val=2.0):
    x_ax = np.arange(n_variables).reshape(-1, 1)
    
    # 1. Kernel-Matrix (K) berechnen
    sq_dist = np.sum(x_ax**2, 1).reshape(-1, 1) + np.sum(x_ax**2, 1) - 2 * np.dot(x_ax, x_ax.T)
    K = np.exp(-0.5 * (1/length_scale**2) * sq_dist) + 1e-8 * np.eye(n_variables)
    
    # --- NEU: KONDITIONIERUNG ---
    # Wir teilen die Matrix auf: K_22 ist die Varianz am fixen Punkt, 
    # K_12 ist die Korrelation des fixen Punkts mit allen anderen.
    K_12 = K[:, const_idx].reshape(-1, 1)
    K_22 = K[const_idx, const_idx]
    
    # Neuer Mittelwert (schiebt alle Kurven zum Constraint)
    mu_cond = K_12 * (1/K_22) * const_val
    mu_cond = mu_cond.flatten()
    
    # Neue Kovarianz (reduziert die Unsicherheit am Constraint auf Null)
    K_cond = K - (1/K_22) * np.dot(K_12, K_12.T)
    # ----------------------------
    
    # Daten generieren mit konditioniertem Mean und Cov
    data = np.random.multivariate_normal(mu_cond, K_cond, n_series)
    
    # Normalisierung fr Farben (basierend auf dem Betrag)
    all_values_abs = np.abs(data)
    dist_min, dist_max = all_values_abs.min(), all_values_abs.max()
    
    fig, ax = plt.subplots(figsize=(12, 6))

    for i in range(n_series):
        y_values = data[i, :]
        x_positions = np.arange(n_variables)
        
        ax.plot(x_positions, y_values, color="black", alpha=0.1, linewidth=0.8, zorder=1)
        
        for j in range(n_variables):
            # Farbe: Dunkel am Mean (0), hell weiter weg
            norm_dist = (np.abs(y_values[j]) - dist_min) / (dist_max - dist_min)
            point_color = cm.RdPu_r(norm_dist)
            ax.scatter(x_positions[j], y_values[j], color=point_color, alpha=0.6, s=20, zorder=2)
            
    # Constraint optisch markieren
    ax.axvline(const_idx, color='black', linestyle='--', alpha=0.3)
    ax.scatter(const_idx, const_val, color='blue', s=100, marker='X', label='Constraint', zorder=5)
    
    ax.set_xlabel('Random Variable', fontsize=12)
    ax.set_ylabel('Observed value', fontsize=12)
    # Adjust ticks based on n_variables
    if n_variables <= 10:
        tick_indices = list(range(n_variables))
    elif n_variables <= 50:
        tick_indices = list(range(0, n_variables, 10))
    else:
        tick_indices = list(range(0, n_variables, 20))
    
    ax.set_xticks(tick_indices)
    ax.set_xticklabels([f'$f(\mathbf{{a}}_{{{i+1}}})$' for i in tick_indices])
    
    # Styling
    ax.set_ylim([-4, 4])
    plt.show()

# Widgets
widgets.interact(plot_gp_constrained,
                 n_variables=widgets.IntSlider(min=20, max=200, value=100),
                 length_scale=widgets.FloatSlider(min=1.0, max=50.0, value=10.0),
                 n_series=widgets.IntSlider(min=1, max=50, value=10),
                 const_idx=widgets.IntSlider(min=0, max=99, value=20, description='Index'),
                 const_val=widgets.FloatSlider(min=-3.0, max=3.0, value=1.5, description='Value'))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import ipywidgets as widgets

def plot_gp_flexible_constraints(n_variables=100, length_scale=10.0, n_series=5, 
                                 count=1, c1_idx=20, c1_val=1.5, 
                                 c2_idx=50, c2_val=-1.0, 
                                 c3_idx=80, c3_val=0.5):
    
    x_ax = np.arange(n_variables).reshape(-1, 1)
    
    # 1. Kernel-Matrix (K)
    sq_dist = np.sum(x_ax**2, 1).reshape(-1, 1) + np.sum(x_ax**2, 1) - 2 * np.dot(x_ax, x_ax.T)
    K = np.exp(-0.5 * (1/length_scale**2) * sq_dist) + 1e-8 * np.eye(n_variables)
    
    # --- DYNAMISCHE CONSTRAINTS AUSWHLEN ---
    all_indices = [c1_idx, c2_idx, c3_idx]
    all_values = [c1_val, c2_val, c3_val]
    
    # Nur die ersten 'count' Constraints nehmen
    const_indices = all_indices[:count]
    const_values = all_values[:count]
    # ----------------------------------------
    
    # Konditionierung
    K_obs_obs = K[np.ix_(const_indices, const_indices)]
    K_all_obs = K[:, const_indices]
    K_obs_inv = np.linalg.inv(K_obs_obs)
    
    mu_cond = (K_all_obs @ K_obs_inv @ np.array(const_values)).flatten()
    K_cond = K - (K_all_obs @ K_obs_inv @ K_all_obs.T)
    K_cond += 1e-10 * np.eye(n_variables)
    
    data = np.random.multivariate_normal(mu_cond, K_cond, n_series)
    dist_min, dist_max = np.abs(data).min(), np.abs(data).max()
    
    fig, ax = plt.subplots(figsize=(12, 6))

    for i in range(n_series):
        y_values = data[i, :]
        ax.plot(np.arange(n_variables), y_values, color="black", alpha=0.1, linewidth=0.8)
        
        for j in range(n_variables):
            norm_dist = (np.abs(y_values[j]) - dist_min) / (dist_max - dist_min) if dist_max > dist_min else 0
            ax.scatter(j, y_values[j], color=cm.RdPu_r(norm_dist), alpha=0.5, s=20)
            
    # Aktive Constraints einzeichnen
    ax.scatter(const_indices, const_values, color='blue', s=120, marker='X', label='Aktiv', zorder=5)
    for idx in const_indices:
        ax.axvline(idx, color='black', linestyle='--', alpha=0.2)

    
    ax.set_xlabel('Random Variable', fontsize=12)
    ax.set_ylabel('Observed value', fontsize=12)
    # Adjust ticks based on n_variables
    if n_variables <= 10:
        tick_indices = list(range(n_variables))
    elif n_variables <= 50:
        tick_indices = list(range(0, n_variables, 10))
    else:
        tick_indices = list(range(0, n_variables, 20))
    
    ax.set_xticks(tick_indices)
    ax.set_xticklabels([f'$f(\mathbf{{a}}_{{{i+1}}})$' for i in tick_indices])
    
    # Styling
    ax.set_ylim([-3, 3])
    
    plt.tight_layout()
    filename = f"funktion_constrained_3_{n_variables}with_dots.png"
    # filename = f"funktion_all_lines{n_variables}.png"
    filepath = os.path.join(save_folder, filename)

    plt.savefig(filepath, dpi=300, bbox_inches='tight')
    plt.show()

# Widget mit Auswahl der Anzahl
widgets.interact(plot_gp_flexible_constraints,
                 n_variables=widgets.fixed(100),
                 length_scale=widgets.FloatSlider(min=1.0, max=50.0, value=15.0),
                 n_series=widgets.IntSlider(min=1, max=20, value=10),
                 count=widgets.IntSlider(min=1, max=3, value=3, description='Anzahl Fixpunkte'),
                 c1_idx=widgets.IntSlider(min=0, max=99, value=15, description='C1 Index'),
                 c1_val=widgets.FloatSlider(min=-3.0, max=3.0, value=1.5, description='C1 Value'),
                 c2_idx=widgets.IntSlider(min=0, max=99, value=50, description='C2 Index'),
                 c2_val=widgets.FloatSlider(min=-3.0, max=3.0, value=-1.0, description='C2 Value'),
                 c3_idx=widgets.IntSlider(min=0, max=99, value=85, description='C3 Index'),
                 c3_val=widgets.FloatSlider(min=-3.0, max=3.0, value=0.5, description='C3 Value'))
```
